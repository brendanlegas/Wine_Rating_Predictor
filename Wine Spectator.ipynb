```python
from bs4 import BeautifulSoup
import pandas as pd
import requests

#URl
wine_url = 'https://top100.winespectator.com/lists/'

#html = requests.get(wine_url).text

# Create BeautifulSoup object; parse with 'html.parser'
#soup = BeautifulSoup(html, 'lxml')
#wine_data_table = soup.find('table')[0]
#wine_data_table

wine_listing_df, = pd.read_html(wine_url,header=0)

wine_listing_df
```

```python
wine_list = wine_listing_df[['Rank', 'Wine', 'Vintage', 'Score', 'Price']]
wine_list.head()
```

```python
#wine_list.dtypes
#print(wine_list['Wine'])

for row in wine_list['Wine']:
    print(str(row))
```

```python
#Run table 'Wine' column through NLP algorithm

from flask import Flask,render_template,url_for,request
import pandas as pd 
import pickle
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.externals import joblib

nlp_model = open('wine_nlp_mod.pkl','rb')
vectorizer = open('vectorizer.pkl','rb')
rfc = joblib.load(nlp_model)
cv = joblib.load(vectorizer)
```

```python
predictor=[]
#for every row in 'Wine'
for row in wine_list['Wine']:
    input_text = str(row)
    data = [input_text]
    vect = cv.transform(data).toarray()
    model_prediction = rfc.predict(vect)
    model_prediction = str(model_prediction)
    model_prediction = model_prediction[1:2]
    predictor.append(model_prediction)
    

predictor
```

```python
wine_list['Model Rating'] = predictor
wine_list
```
