```python
import pandas as pd

df = pd.read_csv('2018.csv')
df
```

```python
clean1 = df[['Wine', 'Vintage', 'Score', 'Price']]
clean1.head()
```

```python
wine_name = []
wine_description = []
vintage = []
score = []
price = []
wine_name.append(clean1['Wine'].iloc[0:199:2])
wine_description.append(clean1['Wine'].iloc[1:200:2])
vintage.append(clean1['Vintage'].iloc[0:199:2])
score.append(clean1['Score'].iloc[0:199:2])
price.append(clean1['Price'].iloc[0:199:2])

vintage
```

```python
winey = pd.DataFrame(wine_name)
transposed = winey.transpose()
transposed.reset_index(inplace=True)
transposed1 = transposed['Wine']

winey = pd.DataFrame(wine_description)
transposed = winey.transpose()
transposed.reset_index(inplace=True)
transposed2 = transposed['Wine']

winey = pd.DataFrame(vintage)
transposed = winey.transpose()
transposed.reset_index(inplace=True)
transposed3 = transposed['Vintage']

winey = pd.DataFrame(price)
transposed = winey.transpose()
transposed.reset_index(inplace=True)
transposed4 = transposed['Price']

winey = pd.DataFrame(score)
transposed = winey.transpose()
transposed.reset_index(inplace=True)
transposed5 = transposed['Score']
```

```python
final_df = pd.DataFrame({
    'Wine': transposed1,
    'Description': transposed2,
    'Vintage': transposed3,
    'Price': transposed4,
    'Score': transposed5
})
final_df.head(100)
```

```python
#Run table 'Wine' column through NLP algorithm

from flask import Flask,render_template,url_for,request
import pandas as pd 
import pickle
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.externals import joblib

nlp_model = open('wine_nlp_model2.pkl','rb')
vectorizer = open('vectorizer2.pkl','rb')
rfc = joblib.load(nlp_model)
cv = joblib.load(vectorizer)
```

```python
predictor=[]
#for every row in 'Wine'
for row in final_df['Description']:
    input_text = str(row)
    data = [input_text]
    vect = cv.transform(data).toarray()
    model_prediction = rfc.predict(vect)
    model_prediction = str(model_prediction)
    model_prediction = model_prediction[1:2]
    predictor.append(model_prediction)
    

predictor
```

```python
final_df['Model Rating'] = predictor
final_df.to_csv('model_2018_2.csv')
final_df
```
